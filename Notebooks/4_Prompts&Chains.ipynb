### Prompts in langgraph

In this tutorial, we will learn how to use prompts in langgraph. We will create a simple graph that takes a topic as input and generates a post and a curated post as output. We will use the ChatOpenAI model to generate the post and the curated post.

```python
from langchain_openai import ChatOpenAI

from dotenv import load_dotenv
import os
load_dotenv()

if os.environ['OPENAI_API_KEY']:
    print("OpenAI API Key is set.")
else:
    raise ValueError("OpenAI API Key is not set.")
```

```python
from langchain_core.messages import HumanMessage

llm = ChatOpenAI(model="gpt-5-mini")
```

```python
from langchain_core.prompts import ChatPromptTemplate


my_prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant that helps to create and curate posts for a social media platform."),
        ("human","I want to create a post about {topic}")
    ]
)
```

#### Chain With Prompts

```python
chain = my_prompt | llm
```

```python
chain.invoke({"topic": "The importance of data privacy in the digital age"})
```
